## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
# Preview Offices Database
cur.execute("""SELECT * FROM offices;""")
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df
```

```python
# Your code here
cur.execute("""SELECT COUNT(employeeNumber) as total_employees, officeCode,city
            FROM employees
            JOIN offices
            USING(officeCode)
            GROUP BY officeCode
            ORDER BY total_employees ASC;""")
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df
```

```python
## Missing the last Boston office, but let's if we can just get the offices with 0 employees
## focus of from should be offices, since the goal is to ID offices using people, instead of before where the goal
## was identifying people using people by office
## not sure why we must use LEFT JOIN for this to work
cur.execute("""SELECT COUNT(e.employeeNumber) as total_employees, o.officeCode,city
            FROM offices AS o
            LEFT JOIN employees AS e 
            USING(officeCode)
            GROUP BY officeCode
            HAVING total_employees = 0;""")
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df
```

```python
cur.execute("""SELECT htmlDescription FROM productlines;""")
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df.head()
```

```python
# Display the htmlDescription and employee's first and last name for each product that each employee has sold
# Your code here
cur.execute("""SELECT htmlDescription, firstName, lastName
            FROM employees e
            JOIN customers c
            ON e.employeeNumber = c.salesRepEmployeeNumber
            JOIN orders o
            USING(customerNumber)
            JOIN orderdetails od
            USING(orderNumber)
            JOIN products p
            USING(productCode)
            JOIN productlines pl
            USING(productLine)""")
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df
```

```python
# Which employees from Boston report to Anthony Bow?
# Your code here
cur.execute("""SELECT firstName, lastName, employeeNumber, reportsTo
            FROM employees
            WHERE reportsTo = 1143;""") ##simplest way is to figure out Anthony Bow's employee Number
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df
```

```python
# Which office got the most business (has highest number of orders)?
# Your code here
cur.execute("""SELECT officeCode, o.city, COUNT(orderNumber)
               FROM offices o
               JOIN employees e
               USING(officeCode)
               JOIN customers c
               ON e.employeeNumber = c.salesRepEmployeeNumber
               JOIN orders o
               USING(customerNumber)
               GROUP BY officeCode
               ORDER BY COUNT(orderNumber) DESC""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print(len(df))
df
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Your code here
cur.execute("""SELECT firstName, lastName, productName
               FROM employees e
               JOIN customers c
               ON e.employeeNumber = c.salesRepEmployeeNumber
               JOIN orders o
               USING(customerNumber)
               JOIN orderdetails od
               USING(orderNumber)
               JOIN products p
               USING(productCode)""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print(len(df))
df.head()
```

## Level Up: Display the Number of Products each employee has sold

```python
# Your code here
cur.execute("""SELECT firstName, lastName, COUNT(productName)
               FROM employees e
               JOIN customers c                   
               ON e.employeeNumber = c.salesRepEmployeeNumber
               JOIN orders o 
               USING(customerNumber)
               JOIN orderdetails od 
               USING(orderNumber)                    
               JOIN products p 
               USING(productCode)
               GROUP BY lastName
               ORDER BY firstName""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print(len(df))
df
```
